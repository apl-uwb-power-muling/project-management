```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = pd.read_csv( "SD_Card/DAT10043.TXT")
print(data)
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(12,12))

data['temp_c'].plot( ax=axes[0])
data['temp_from_pressure_c'].plot( ax=axes[0])
axes[0].set_xlabel('Time (s)')
axes[0].set_ylabel('Temperature (deg C)')

data['pressure_mbar'].plot(ax=axes[1], ylim=(1000,1100))
axes[1].set_ylabel('Pressure (mbar)')

data['depth_m'].plot(ax=axes[2], ylim=(1.0,-0.5))
axes[2].set_ylabel('Depth (m)')
```
